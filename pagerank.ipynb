## Pagerank example (from lecture notes "Social Network Analysis I")

Example 1: Consider the following graph. 

```python
import networkx as nx
import matplotlib.pyplot as plt 

G = nx.DiGraph()
G.add_nodes_from(['a','y','m'])
G.add_edge('a', 'y')
G.add_edge('y', 'a')
G.add_edge('y', 'y')
G.add_edge('a', 'm')
G.add_edge('m', 'a')

nx.draw_spectral (G,node_size=800,with_labels=True, node_color='y')
plt.show()
```

Note that in  networkx, the self-loop from y to y is not drawn.

We can compute the pagerank by power iteration

```python
from numpy  import *
x=[1/3 , 1/3, 1/3] # [y, m, a]
M = [[1/2,1/2,0],
     [1/2,0,1],
     [0,1/2,0]] #probability matrix
r = transpose(x) #initial pagerank


for i in range(0, 100):
    print("iteration:%d"%i)
    print(r)
    r =matmul(M,r)
```

Example 2: Consider the following graph


```python
import networkx as nx
import matplotlib.pyplot as plt 

G = nx.DiGraph()
G.add_nodes_from(['a','y','m'])
G.add_edge('a', 'y')
G.add_edge('a', 'm')
G.add_edge('y', 'm')
G.add_edge('m', 'y')

nx.draw_spectral (G,node_size=800,with_labels=True, node_color='y')
plt.show()
```

Note that y and m is equally important and have the same page rank(0.5). However, node a has no importance (as there are no incoming edges)

```python
from numpy  import *
x=[1/3 , 1/3, 1/3] # [y, m, a]
M = [[0,1,1/2],
     [1,0,1/2],
     [0,0,0]] #probability matrix
r = transpose(x) #initial pagerank


for i in range(0, 20):
    print("iteration:%d"%i)
    print(r)
    r =matmul(M,r)
```

## Pagerank with damping factor d

```python
from numpy  import *
x=[1/3 , 1/3, 1/3] # [y, m, a]
M = [[0,1,1/2],
     [1,0,1/2],
     [0,0,0]] #probability matrix
r = transpose(x) #initial pagerank
R = [[1/3,1/3,1/3],
     [1/3,1/3,1/3],
     [1/3,1/3,1/3]] #matrix to add randomness

d = 0.8 #damping factor
M2 = add(multiply(R,(1-d)), multiply(M,d))
print (M2)
```

```python
for i in range(0, 100):
    print("iteration:%d"%i)
    print(r)
    r =matmul(M2,r)
```

Note that the importance of y and m is slightly decrease to 0.467, but a still have some importance.
We may cross check the result with networkx pagerank function.

```python
 pr = nx.pagerank(G, alpha=0.8)
print(pr)
```
