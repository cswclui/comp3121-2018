```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph()
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(1,4)
G.add_edge(2,3)
G.add_edge(3,4)
G.add_edge(4,5)
G.add_edge(4,6)
G.add_edge(5,6)
G.add_edge(5,7)
G.add_edge(6,7)
G.add_edge(5,8)
G.add_edge(6,8)
G.add_edge(7,8)
G.add_edge(7,9)


nx.draw(G, with_labels=True)
plt.show()
```

## Find edge with highest betweeness centrality


```python

eb=nx.edge_betweenness_centrality(G, normalized=False)
print(eb)


def get_max_betweeness_edge():
    max = 0
    max_edge = ()
    for k, v in eb.items():
        if v > max:
            max = v 
            max_edge=k
    return max, max_edge

(max,max_edge) = get_max_betweeness_edge()

print(max)
print(max_edge)
        
```

## Remove edge with highest between centrality


```python
G.remove_edge(max_edge[0],max_edge[1])
nx.draw(G, with_labels=True)
plt.show()
```

## Find edge with highest betweeness centrality in the new graph and delete the edge

```python
eb=nx.edge_betweenness_centrality(G, normalized=False)
print(eb)

(max,max_edge) = get_max_betweeness_edge()
print(max)
print(max_edge)
G.remove_edge(max_edge[0],max_edge[1])
nx.draw(G, with_labels=True)
plt.show()
      
```

## Find edge with highest betweeness centrality in the new graph and delete the edge

```python
eb=nx.edge_betweenness_centrality(G, normalized=False)
print(eb)

(max,max_edge) = get_max_betweeness_edge()
print(max)
print(max_edge)
G.remove_edge(max_edge[0],max_edge[1])
nx.draw(G, with_labels=True)
plt.show()
```
